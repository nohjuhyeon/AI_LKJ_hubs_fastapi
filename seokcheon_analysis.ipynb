```python
from konlpy.tag import Okt
okt = Okt()

def remove_stopwords(text, stopwords):
    result = [word for word in text if word not in stopwords]
    return result

# ì¡°ì‚¬, ì–´ë¯¸ êµ¬ë‘ì  ê°™ì€ ê²ƒì„ ì œê±° í•¨ìˆ˜ / ë‚˜ì¤‘ì— ë¶ˆìš©ì–´ ì²˜ë¦¬ í•¨ìˆ˜ë„ ì‘ì„±í•´ì£¼ëŠ” ê²ƒì´ ì¢‹ìŒ
def okt_clean(text):
    clean_text = []
    for word in okt.pos(text, stem=True):
        if word[1] not in ['Josa', 'Eomi', 'Punctuation','Determiner','Conjunction']:
            clean_text.append(word[0])
    # ë¶ˆìš©ì–´ ì²˜ë¦¬
    stopwords = ['ê°€', 'ê³ ', 'ì„', 'ë¥¼', 'ì´', 'ëŠ”']  # ë¶ˆìš©ì–´ ë¦¬ìŠ¤íŠ¸
    clean_text = remove_stopwords(clean_text, stopwords)
    return " ".join(clean_text)
```

```python
df_title["card_content"] = df_title["card_content"].apply(okt_clean)
df_title['card_date'] = pd.to_datetime(df_title['card_date'], format='%Yë…„ %mì›” %dì¼')
df_title
```

```python
df_title['card_year'] = pd.to_datetime(df_title['card_date']).dt.year
df_title['card_month'] = pd.to_datetime(df_title['card_date']).dt.month
# ê³„ì ˆ ì»¬ëŸ¼ ì¶”ê°€
def get_season(month):
    if month in [3, 4, 5]:
        return 'ë´„'
    elif month in [6, 7, 8]:
        return 'ì—¬ë¦„'
    elif month in [9, 10, 11]:
        return 'ê°€ì„'
    else:
        return 'ê²¨ìš¸'

df_title["season"] = df_title['card_month'].apply(get_season)
df_title
```

```python
df_spring = df_title.query("season == 'ë´„'")
df_summer = df_title.query("season == 'ì—¬ë¦„'")
df_autumn = df_title.query("season == 'ê°€ì„'")
df_winter = df_title.query("season == 'ê²¨ìš¸'")

list_spring = df_spring["card_content"].to_list()
list_summer = df_summer["card_content"].to_list()
list_autumn = df_autumn["card_content"].to_list()
list_winter = df_winter["card_content"].to_list()
```

```python
# ë¶ˆìš©ì–´ ë¦¬ìŠ¤íŠ¸ ìƒì„± (ì˜ˆì‹œ)
stopwords = ['ê°€', 'ê³ ', 'ì„', 'ë¥¼', 'ì´', 'ëŠ”']
from konlpy.tag import Okt

# Okt í˜•íƒœì†Œ ë¶„ì„ê¸° ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
okt = Okt()

# í† í¬ë‚˜ì´ì§• í•¨ìˆ˜ ì •ì˜
def tokenizer(raw, pos=["Noun","Alpha","Verb","Number"], stopword=stopwords):
    return [
        word for word, tag in okt.pos(
            raw, 
            norm=True,   
            stem=True    
            )
            if len(word) > 1 and tag in pos and word not in stopword

    ]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=tokenizer, use_idf=True)
features = vectorizer.fit_transform(list_spring)
features.toarray()
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_model = LatentDirichletAllocation(n_components=5, random_state=55) #ì¸ìŠ¤í„´ìŠ¤í™” #n_components í† í”½ì˜ ê°¯ìˆ˜
lda_model.fit(features) #êµìœ¡
```

```python
dictionary_list = vectorizer.get_feature_names_out()
lda_model.components_
```

```python
topics_output = lda_model.transform(features)
df_topics_score = pd.DataFrame(data=topics_output)
df_topics_score['dominant_topic_number'] = np.argmax(topics_output, axis=1) #  "topics_output" ë°°ì—´ì—ì„œ ê° í–‰ë§ˆë‹¤ ìµœëŒ€ê°’ì˜ ì¸ë±ìŠ¤ë¥¼ ì°¾ì•„ "dominant_topic_number" ì—´ì— í• ë‹¹
df_topics_score['sentences'] = list_spring
df_topics_score
```

```python
## ìƒìœ„ ë‹¨ì–´ ì¶”ì¶œ
topics_list = list()
for topic in lda_model.components_:
    df_datas = [topic, dictionary_list]
    df_topics = pd.DataFrame(data=df_datas)
    df_topics= df_topics.T #TëŠ” í–‰ê³¼ ì—´ì„ ë°”ê¿ˆ
    df_topics = df_topics.sort_values(0, ascending=False)
    # print(df_topics[:3])
    topics_text = ' '.join(df_topics[1].values[:10])# ì‹œë¦¬ì¦ˆ í˜•ì‹ìœ¼ë¡œ ì¶œë ¥ get values from series / index
    print(topics_text)
    topics_list.append(topics_text)
topics_list_add = [topics_list, ['Topic0', 'Topic1', 'Topic2', 'Topic3', 'Topic4']]
df_topics_spring = pd.DataFrame(topics_list_add)
df_topics_spring
```
